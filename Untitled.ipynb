```python
def sniff_modified(df):
    with pd.option_context("display.max_colwidth", 20):
        info = pd.DataFrame()
        info['data type'] = df.dtypes
        info['percent missing'] = df.isnull().sum()*100/len(df)
        info['No. unique'] = df.apply(lambda x: len(x.unique()))
        info['unique values'] = df.apply(lambda x: x.unique())
        return info.sort_values('data type')
sniff_modified(X_num)
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X_num = pd.DataFrame(st.fit_transform(X_num), columns = X_num.columns)
X_num
```

```python
from pandas.api.types import is_object_dtype
obj_cols = []
for col in X.columns:
    if is_object_dtype(X[col]):
        obj_cols.append(col)
X = pd.concat([X_num, data[obj_cols]], axis = 1)
sniff_modified(X)
```

```python
X = X.fillna(0)
```

```python
import category_encoders as ce
encoder = ce.SumEncoder(cols = obj_cols)
encoder.fit(X)
X = encoder.transform(X)
X
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
import tensorflow as tf
stop = EarlyStopping(monitor='val_loss', mode = min, patience = 200)
model = Sequential()
model.add(Dense(256, input_dim = 262, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))
model.compile(optimizer = 'sgd', loss = 'msle', metrics = [tf.keras.metrics.MeanSquaredLogarithmicError()])
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', verbose=1, save_best_only=True)
model.summary()
```

```python
history = model.fit(X, y, epochs = 4000, validation_split = 0.1, callbacks = [stop, mc], batch_size = 8)
```

```python
saved_model = load_model('best_model.h5')
```

```python
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_log_error
scorer = make_scorer(mean_squared_log_error, greater_is_better=False, size=10)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
params = {
 'learning_rate' : [0.05,0.10,0.01],
 'max_depth' : [ 8, 10],
 'min_child_weight' : [ 1, 3],
 'gamma': [ 0.0, 0.1, 0.2 ],
 'colsample_bytree' : [ 0.3, 0.5 , 0.7 ],
 'subsample' : [ 0.5, 1],
 'reg_alpha' : [0.1, 0.5, 1],
 'reg_lambda' : [0.1, 0.5, 1]
}
rs_model=GridSearchCV(xgb.XGBRegressor(),param_grid=params,scoring=scorer,n_jobs=-1,cv=3,verbose=3)
rs_model.fit(X, y_log)
xg1= rs_model.best_estimator_
```
